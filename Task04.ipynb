```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    if plan not in valid_plans:
        print("Invalid plan selected. Available plans: 199, 299, 399, 599.")
        return False
    if amount != plan:
        print("Recharge amount must match the selected plan.")
        return False
    return True
# Main program with retry option
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        plan = int(input("Enter plan (199, 299, 399, 599): "))
        if validate_recharge(amount, plan):
            print("Recharge successful!")
            break
        else:
            print("Please try again.\n")
    except ValueError:
        print("Please enter valid numeric values.\n")
```

```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")
# Sample dictionary of products and stock quantities
inventory = {
    "Rice": 25,
    "Wheat": 10,
    "Sugar": 18,
    "Oil": 8,
    "Salt": 30
}
# Function call
check_stock(inventory)
```

```python
def evaluate_student(marks):
    total = 0
    # Loop to calculate total marks
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"
# Example usage
student_marks = [60, 45, 55, 70, 40]
result = evaluate_student(student_marks)
print(result)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    if peak.lower() == "yes":
        fare *= 1.25   # 25% extra during peak hours
    return fare
# Main program with retry option
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")
        
        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using the Cab Fare Calculator!")
            break
    except ValueError:
        print("Please enter valid numeric distance.\n")
```

```python
def check_eligibility(attendance):
    present_days = 0
    # Loop to count present days
    for status in attendance:
        if status == "P":
            present_days += 1
    total_days = len(attendance)
    percentage = (present_days / total_days) * 100
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"
# Example usage
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance_list)
print(result)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"
    # Loop over string
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    # Conditions check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return " Weak Password (Must contain at least one digit and one special character @#$)"
# Example usage
user_password = input("Enter your password: ")
result = check_password_strength(user_password)
print(result)
```
