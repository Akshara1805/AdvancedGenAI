```python
# Input dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Identify the highest performance score
highest_score = max(employees.values())
# Find all employees with the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]
# Display the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
import string
# Input sentence
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))
# Split sentence into words
words = query.split()
# Count frequency of each word
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
# Display only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

```python
# Input list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Identify even readings and store as (hour_index, reading_value)
valid_readings = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even
        valid_readings.append((index, value))
# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Input list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# Count users per domain
domain_count = {}
for email in emails:
    domain = email.split("@")[1]  # Extract domain
    domain_count[domain] = domain_count.get(domain, 0) + 1
# Calculate percentage usage
total_users = len(emails)
print("Email Domain Usage:")
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Input list
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Calculate average daily sales
average_sales = sum(sales) / len(sales)
# Determine threshold (30% above average)
threshold = average_sales * 1.30
# Detect sales spikes
print("Sales Spikes Detected:")
for index, value in enumerate(sales, start=1):  # start=1 for day number
    if value > threshold:
        print(f"Day {index}: {value}")
```

```python
# Input list
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count occurrences of each user ID
id_count = {}
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1
# Display only duplicates (count > 1)
print("Duplicate User IDs:")
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
