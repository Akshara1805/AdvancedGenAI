```python
### Unique words in a sentance
```

```python
# Take unique words
sentance = input("Enter sentance:")
# Split the sentence
words = sentance.split()
# count the words
unique_words = set(words)
# print 
print("Count Unique words:", len(unique_words))
print("Unique words:", unique_words)
```

```python
### Highest salary from Employee data
```

```python
# dictionary
marks = {"Ravi": 75000 , "Anita": 68000 , "Kiran": 72000}
# Assume first employee as highest 
first_key = list(marks.keys())[0]
highest = marks[first_key]
highest_name = first_key
# loop
for name,salary in marks.items():
    if salary > highest:
        highest = salary
        highest_name = name
print("Highest salary:", highest_name,"-",highest)
```

```python
### Find Maximum and Minimum Values
```

```python
# list
numbers = [45, 22, 89, 10, 66]
# initialize min and max
max = numbers[0]
min = numbers[0]
# for loop
for num in numbers:
    if num > max:
        max = num
    if num < min:
        min = num
# Print
print("List:", numbers)
print("Maximum numbers:", max)
print("Minimun numbers:", min)
```

```python
### Count products Above a Price Threshold
```

```python
# list
prices = [450, 1200, 899, 1500, 300]
#initialize count
count = 0
# for loop
for price in prices:
    if price > 1000:
        count = count + 1
print("Products above 1000:", count)
```

```python
### Calculate Attendance Persentage
```

```python
# list
attendance = ["P","P","A","P","P"]
# Count
count = 0
# for loop
for status in attendance:
    if status == "P":
        count = count + 1
# Calculate Percentage
total_days = len(attendance)
percentage = (count / total_days) * 100
# print
print("Attendance Percentage:", percentage)
```

```python
### Count character frequency
```

```python
# dictionary
text = "ppython"
# empty dictionary
char = {}
# for loop
for ch in text:
    if ch in char:
        char[ch] += 1
    else:
        char[ch] = 1
print(char)
```
