```python
correct_username = "admin"
correct_password = "1234"
# Taking input from use
username = input("Enter username: ")
password = input("Enter password: ")
if username == correct_username and password == correct_password:
    print("Login successful")
else:
    print("Invalid")
```

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
# Checking each student's marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Students :", len(marks))
print("Total Pass Students :", pass_count)
print("Total Fail Students :", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
# Clean and standardize names
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
# Analyze each message
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)
    if length > 10:
        print("Flag: Message is longer than 10 characters")
    print()
```
