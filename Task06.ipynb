```python
def manage_parking(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0
    for log in vehicle_logs:
        if log == 'IN':
            current_parked += 1
        elif log == 'OUT' and current_parked > 0:
            current_parked -= 1
        # peak parking
        if current_parked > peak_usage:
            peak_usage = current_parked
        # check status
        if current_parked > capacity:
            status = "ALERT"
        else:
            status = 'Available'
        # output
        print('Currently Parked Vehicles:', current_parked)
        print('Parking status:',status)
# input
capacity = 50
vehicle_logs = ['IN','IN','OUT','IN','IN','OUT']
manage_parking(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(dis,traffic,weather):
    # base 
    base_time = (dis/30)*60
    delay =0
    # traffic delay
    if traffic == 'High':
        delay += 25
    elif traffic == 'Medium':
        delay += 5
    elif traffic == "Low":
        delay += 0
    # weather delay
    if weather == "Rainy":
        delay += 14
    elif weather == "Stromy":
        delay += 25
    elif weather == "Clear":
        delay += 0
    total = int(base_time+delay)
    print("Estimate Delivery Time:",total,"minutes")
# input
dis = 8
traffic = "High"
weather = "Rainy"
estimate_delivery_time(dis,traffic,weather)
```

```python
def classifier_server_load(cpu_reading):
    total = 0
    for usage in cpu_reading:
        total += usage
    avg = int(total/len(cpu_reading))
    if avg < 50:
        status = "Normal"
    elif 50 <= avg <= 80:
        status = "Warming"    
    else:
        status = 'Critical'
    # output
    print("Average CPU Load:",avg,"%")
    print("Server Status:",status)
# input
cpu_reading = [45,60,70,85,90]
classifier_server_load(cpu_reading)
```

```python
def analyze_resource_usage(resource_usage):
    overused=[]
    for resource,hours in resource_usage.items():
        if hours>6:
            overused.append(resource)
            if overused:
                print("Overused Resources:",",".join(overused))
                print("Enery Alert:YES")
            else:
                print("Overused Resources:None")
                print("Enery Alert:NO")
resource_usage={
    "projector":6,
    "AC":9,
    "Lights":4
}
analyze_resource_usage(resource_usage)
```

```python
def manage_event_registration(capacity,registrations):
    confirmed=0
    waitlisted=0
    for _ in range(registrations):
        if confirmed<capacity:
            confirmed += 1
        else:
            waitlisted += 1
    if confirmed == capacity:
        status = "closed"
    else:
        status = "Open"
    #ouput
    print("Confrimed Registrations:",confirmed)
    print("Waitlisted Users:",waitlisted)
    print("Registration Status:",status)
# output
event_capacity = 100
total_registrations = 105
manage_event_registration(event_capacity,total_registrations)
```
