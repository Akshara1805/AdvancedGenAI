```python
# Smart Light Controller using OOPS
class SmartLight:
    def __init__(self, light_name):
        # initialize the light name
        self.light_name = light_name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(f"{self.light_name} is {self.status}")
# Input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ").upper()
# Create
light = SmartLight(light_name)
#action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off
else:
    print("Invalid")
#Display
light.display_status()
```

```python
# Employee ID Card using OOPS
class Employee:
    def __init__(self, name, empid, dept):
        self.name = name
        self.empid = empid
        self.dept = dept
    def display_id_card(self):
        print(f"Employee ID Card")
        print(f"Name:{self.name}")
        print(f"ID:{self.empid}")
        print(f"Department:{self.dept}")
#input
name=input("Employee Name:")
empid=input("Employee ID:")
dept=input("Department:")
#create object
employee = Employee(name, empid, dept)
#display
employee.display_id_card()
```

```python
# Mobile Contact Record using OOPS
class Contact:
    def __init__(self, name, phno):
        self.name = name
        self.phno = phno
    def display_contact(self):
        print(f"Contact Saved")
        print(f"Name:{self.name}")
        print(f"Phone Number:{self.phno}")
#input
name = input("Enter the name")
phno = input("Phone Number")
#Create object
contact = Contact(name, phno)
#display
contact.display_contact()
```

```python
# Product Price Target Generator using OOPS
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display_price(self):
        print(f"Name:{self.name}")
        print(f"Price:₹{self.price}")
#Input
name=input("Product Name:")
price=input("Price")
#create Object
product=Product(name,price)
#Display
product.display_price()
```

```python
#Movie Rating Display System using OOPS
class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def display_movie(self):
        print(f"Movie:{self.name}")
        print(f"Rating:{self.rating}/5")
#input
name=input("Movie:")
rating=input("Rating:")
#create object
movie=Movie(name,rating)
#Display
movie.display_movie()
```

```python
#Delivery Address Manager using OOPS
class Delivery:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def display_delivery(self):
        print(f"Delivery Details")
        print(f"Customer Name:{self.name}")
        print(f"Address:{self.address}")
# input
name=input("Customer Name: ")
address=input("Address: ")
#create object
delivery=Delivery(name,address)
#Display
delivery.display_delivery()
```
