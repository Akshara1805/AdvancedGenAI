```python
# Agriculture Rainfall Adequacy Checker
def analyze_seasonal(rain_fall, required_level):
    total = 0
    # for loop
    for value in rain_fall:
        # Total
        total += value
        # Average
        avg = total / len(rain_fall)
        # if loop
        if avg >= required_level:
            status = 'Adequite Rainfall'
        else:
            status = 'Inadequite Rainfall'
        return total, avg, status
daily_rain = [70, 75, 68, 74, 73]
required_level = 70
total, avg, result = analyze_seasonal(daily_rain, required_level)
# Print
print("Total:", total)
print("Average Rainfall:", avg)
print("Rainfall status:", result)
```

```python
# Social Media - Duplicate Account Detection
def duplicate_users(user_names):
    unique_users = set(user_names)
    # if condition
    if len(user_names) != len(unique_users):
        status = 'Yes'
    else:
        status = 'No'
    # output
    print("Dupicate Account Found:",status)
user = ['vyshu123', 'Lucky456', 'Uma789', 'Lucky456'] 
# result
result = duplicate_users(user)
```

```python
# E-Learning - Student Attendance Eligibility System 
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    # Loop to count attended classes (1 = present, 0 = absent)
    for day in attendance_list:
        if day == 1:
            attended_classes += 1
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    # output
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)
attendane = [1,1,0,1,1]
check_exam_eligibility(attendane)
```

```python
# Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
units_consumed = 250
bill_amount, status = calculate_electricity_bill(units_consumed)
# output
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", bill_amount)
print("Usage Status:", status)
```
