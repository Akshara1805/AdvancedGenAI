```python
import re
import emoji
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
# Initialize tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # 3. Remove emojis
    text = emoji.replace_emoji(text, replace='')
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    # 5. Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # 6. Tokenization
    tokens = word_tokenize(text)
    # 7. Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    # 8. Stemming
    stemmed = [stemmer.stem(word) for word in tokens]
    # 9. Lemmatization
    lemmatized = [lemmatizer.lemmatize(word) for word in tokens]
    return {
        "clean_text": text,
        "tokens": tokens,
        "stemmed": stemmed,
        "lemmatized": lemmatized
    }
# Example messy text
text = "I LOVE NLP!!! Visit https://example.com now. Running, runs, ran..."
result = preprocess_text(text)
print("Clean Text:", result["clean_text"])
print("Tokens:", result["tokens"])
print("Stemmed:", result["stemmed"])
print("Lemmatized:", result["lemmatized"])
```

```python
import re
# -------------------------------------------------
# Preprocessing Function
# -------------------------------------------------
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    cleaned_text = " ".join(tokens)
    return tokens, cleaned_text
# -------------------------------------------------
# Test Sentences
# -------------------------------------------------
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
# -------------------------------------------------
# Token Analytics
# -------------------------------------------------
print("\n========== TOKEN ANALYTICS ==========\n")
max_noise = 0
max_meaningful = 0
noisiest_sentence = ""
best_sentence = ""
for sentence in test_sentences:
    original_tokens = sentence.split()
    tokens, cleaned = preprocess_text(sentence)
    # Metrics
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0
    # Noise calculation (how much removed)
    noise = len(original_tokens) - total_tokens
    if noise > max_noise:
        max_noise = noise
        noisiest_sentence = sentence
    if total_tokens > max_meaningful:
        max_meaningful = total_tokens
        best_sentence = sentence
    # Print per sentence
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-----------------------------------")
# -------------------------------------------------
# Final Analysis Answers
# -------------------------------------------------
print("\n========== ANALYSIS ==========")
print("Sentence with most noise:")
print(noisiest_sentence)
print("\nSentence with most meaningful tokens:")
print(best_sentence)
```

```python
from collections import Counter
# -----------------------------------------
# Collect all tokens from all sentences
# -----------------------------------------
all_tokens = []
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)
# -----------------------------------------
# Count frequency
# -----------------------------------------
word_freq = Counter(all_tokens)
# Top 10 most frequent words
top_10 = word_freq.most_common(10)
# Top 5 least frequent words
least_5 = word_freq.most_common()[:-6:-1]
# -----------------------------------------
# Print Results
# -----------------------------------------
print("\n========== FREQUENCY ANALYSIS ==========\n")
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word} : {count}")
print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word} : {count}")
```

```python
import re

def preprocess_text(text):
    
    # -------------------------------
    # 1. Handle empty input
    # -------------------------------
    if not text or text.strip() == "":
        return [], ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove punctuation & emojis
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenization
    tokens = text.split()
    # Remove short tokens (≤2 except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    # -------------------------------
    # 2. Handle cases with no tokens left
    # -------------------------------
    if not tokens:
        return [], ""
    cleaned_text = " ".join(tokens)   
    return tokens, cleaned_text
test_cases = [
    "",                     # Empty string
    "😍😍🔥🔥",              # Only emojis
    "1234567890",           # Only numbers
    "!!!???",               # Only symbols
    "   "                   # Only spaces
]
for text in test_cases:
    tokens, cleaned = preprocess_text(text)
    print("\nInput:", repr(text))
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```
